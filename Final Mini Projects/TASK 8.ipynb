```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as smd
df_activities = pd.read_csv("cardio.csv",index_col = "Date", parse_dates=True)
```

```python
df_run_dist_wkly = df_activities.loc["20190101":"20130101"]["Distance (km)"].resample("W").bfill()
fig = plt.figure(figsize=(12,5))
ax = df_run_dist_wkly.plot(label="Trend", linewidth=1.5)
ax.legend()
ax.set_title("Running Trend")
plt.show()
```
