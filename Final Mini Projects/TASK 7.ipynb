```python
import pandas as pd
import matplotlib.pyplot as plt
df_activities = pd.read_csv("cardio.csv",index_col = "Date", parse_dates=True)
df_activities.head()
```

```python
runs_subset_2013_2018 = df_activities.loc["20190101":"20130101"]
df_run_dist_annual = runs_subset_2013_2018["Distance (km)"].resample("A").sum()
```

```python
fig = plt.figure(figsize=(8,5))
# plotting data 
ax = df_run_dist_annual.plot(marker=".", markersize=14, linewidth=0, color="blue")
ax.set(ylim=[0,800],xlim=["2012","2018"],ylabel="Distance (km)", xlabel="years", title="Distance in Years")
ax.axhspan(0,800, color="red", alpha=0.2)
plt.show()
```
