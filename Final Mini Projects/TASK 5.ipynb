```python
import pandas as pd
import matplotlib.pyplot as plt
df_activities = pd.read_csv("cardio.csv",parse_dates=True,index_col="Date")
df_activities.drop(["Route Name", "Friend's Tagged"], inplace=True, axis=1)
```

```python
runs_subset_2015_2018 = df_activities.loc["20190101":"20150101"]     #subsetting from 2015-2018
```

```python
df_run=df_activities.loc[df_activities['Type'] == 'Running'].copy()
runs_subset_2015_2018 = df_run.loc["20181111 14:05:12":"20150102 18:35:11"] #subset from 15-18
runs_subset_2015_2018.plot(subplots=True,figsize=(6,4), marker="*", markersize=8)
plt.show()
```

```python
# Annual Stats-

print('Last 4 years Annual Average stats:')
display(runs_subset_2015_2018.resample("A").mean())
```

```python
print("Weekly stats of last 4 years")
display(runs_subset_2015_2018.resample("W").mean().mean())
```

```python
weekly_counts_average = runs_subset_2015_2018["Distance (km)"].resample("W").count().mean()
weekly_counts_average
```
